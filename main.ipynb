```python
%pylab inline
from __future__ import division

import os
import sys

from linac import Linac, Simulation
```

```python
sim_name = "multiTubeIssue"
runid = os.getpid()

data_directory = "histograms"
run_name = "basic"
```

```python
linac = Linac("machine/asymmetricProblem.yaml")

sim = Simulation(sim_name, linac, run_id=runid)
sim.setup_phantom()
# sim.show()
```

```python
sim.beam_on(int(1e3))
# sim.beam_on(int(50))
```

```python
sim.save_histograms(data_directory, run_name, runid)
```
