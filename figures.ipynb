```python
# Defining sensitive detector dimensions
# as per linac/g4/src/SensitiveDetector.cc

mm = 1

x_dim = 101
y_dim = 101
z_dim = 100

x_min = -75.75 * mm
y_min = -75.75 * mm
z_min = -50 * mm

x_max = 75.75 * mm
y_max = 75.75 * mm
z_max = 0 * mm

x_res = (x_max - x_min) / (x_dim)
y_res = (y_max - y_min) / (y_dim)
z_res = (z_max - z_min) / (z_dim)

x = arange(x_min,x_max,x_res) + x_res/2
y = arange(y_min,y_max,y_res) + y_res/2
z = -(arange(z_min,z_max,z_res) + z_res/2)
```

```python
sum_profile = sum(energy, axis=2)

contourf(x,y,transpose(sum_profile),50)

title("Total energy for all depths in water phantom")
axis('equal')
xlabel("x")
ylabel("y")
colorbar();
```

```python
energy = load("%s/energy_%s_%s_%s.npy" % (data_directory, sim_name, run_name, runid))
```

```python
vertical_slice_dx = sum(sum(energy,axis=1),axis=1)

plot(x,vertical_slice_dx,'b')

title("Total energy of vertical slices moving along x-axis")
xlabel("x")
ylabel("Total energy through vertical slices")

grid(True)
```

```python
vertical_slice_dy = sum(sum(energy,axis=0),axis=1)

plot(y,vertical_slice_dy,'g')

title("Total energy of vertical slices moving along y-axis")
xlabel("y")
ylabel("Total energy through vertical slices")
grid(True)
```

```python
plot(x,vertical_slice_dx,'b')
plot(y,vertical_slice_dy,'g')

title("Overlay")
ylabel("Total energy through vertical slices")
xlabel("x | y")

grid(True)
```
